```python
from langchain_community.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings import OpenAIEmbeddings
from langchain.chat_models import ChatOpenAI
from langchain.vectorstores import Pinecone
import pinecone
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import pipeline
from langchain import HuggingFacePipeline
from huggingface_hub import notebook_login
import textwrap
import sys
import os
import torch
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
urls=['https://www.arbabmustafa.com/',
      'https://devvibe.com/']
```

```python
loaders=UnstructuredURLLoader(urls=urls)
data=loaders.load()
```

```python
data
```

```python
len(data)
```

```python
text_splitter=CharacterTextSplitter(
    separator='\n',
    chunk_size=1000,
    chunk_overlap=200
)
```

```python
text_chunks=text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
#dowanload hugging face embeddings
```

```python
embeddings=HuggingFaceEmbeddings()
```

```python
embeddings
```

```python
from langchain_community.vectorstores import FAISS
```

```python
!pip install faiss-cpu
```

```python
vector_store = FAISS.from_documents( text_chunks, embeddings)
```

```python
from langchain.chains import RetrievalQAWithSourcesChain
```

```python
model = "daryl149/llama-2-7b-chat-hf"
```

```python
notebook_login()
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
!pip install -U bitsandbytes
```

```python
model = AutoModelForCausalLM.from_pretrained(model,
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              #load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
#it will un only on gpu T4
```

```python
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 50,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.predict("Please provide a concise summary of the Book Harry Potter") #checking llm output
```

```python
from langchain.chains import RetrievalQA
```

```python
query = "what is flutter?"
```

```python
docs = vector_store.similarity_search(query, k=3)
```

```python
docs
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vector_store.as_retriever())
```

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vector_store.as_retriever())
```

```python
query="what is flutter?"
qa.run(query)
```

```python
query2="where Devvibe is located?"
qa.invoke(query2)
```
